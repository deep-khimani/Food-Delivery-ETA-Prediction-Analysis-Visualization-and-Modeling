```python
print(train_data.dtypes)
for col in train_data.columns:
    print(f"{col}: {train_data[col].nunique()} unique values")
```

```python
categorical_cols = [
    'Weatherconditions', 'Road_traffic_density', 'Vehicle_condition',
    'Type_of_order', 'Type_of_vehicle', 'Festival', 'City'
]
for col in categorical_cols:
    print(train_data[col].value_counts())
    sns.countplot(data=train_data, x=col, order=train_data[col].value_counts().index)
    plt.title(f'Distribution: {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
numeric_cols = [
    'Delivery_person_Age', 'Delivery_person_Ratings', 'Restaurant_latitude',
    'Restaurant_longitude', 'Delivery_location_latitude',
    'Delivery_location_longitude', 'Vehicle_condition',
    'multiple_deliveries', 'Time_taken(min)'
]
train_data[numeric_cols].hist(bins=20, figsize=(15,10))
plt.tight_layout()
plt.show()

# Summary statistics
print(train_data[numeric_cols].describe())
```

```python
train_data['Order_Date'] = pd.to_datetime(train_data['Order_Date'], errors='coerce')
print(train_data['Order_Date'].min(), train_data['Order_Date'].max())

# Analyze ordering times (if 'Time_Orderd' is valid)
print(train_data['Time_Orderd'].head())

# Extract hour and weekday for further EDA
train_data['Order_Hour'] = pd.to_datetime(train_data['Time_Orderd'], errors='coerce').dt.hour
train_data['Order_Weekday'] = train_data['Order_Date'].dt.day_name()
sns.countplot(data=train_data, x='Order_Hour')
plt.title('Order Distribution by Hour')
plt.show()
```

```python
print(train_data[['Restaurant_latitude', 'Restaurant_longitude',
          'Delivery_location_latitude', 'Delivery_location_longitude']].describe())
```
